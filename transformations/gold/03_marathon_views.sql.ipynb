```python
%sql
-- Databricks notebook source

-- MAGIC %md
-- MAGIC # Gold — marathon views
-- MAGIC Two simple views based on race distance unit: km and mi.

-- COMMAND ----------

CREATE OR REPLACE VIEW marathos.gold.v_marathon_km_results AS
SELECT
  f.result_id,
  e.event_name,
  e.event_year,
  a.athlete_country,
  a.athlete_gender,
  f.performance_seconds,
  f.distance_km,
  f.speed_kmh,
  ROUND(f.performance_seconds / 3600.0, 2) AS finish_hours
FROM marathos.gold.fct_results f
JOIN marathos.gold.dim_event e
  ON f.event_id = e.event_id
JOIN marathos.gold.dim_athlete a
  ON f.athlete_id = a.athlete_id
WHERE e.distance_unit = 'km';

-- COMMAND ----------

CREATE OR REPLACE VIEW marathos.gold.v_marathon_mi_results AS
SELECT
  f.result_id,
  e.event_name,
  e.event_year,
  a.athlete_country,
  a.athlete_gender,
  f.performance_seconds,
  f.distance_km,
  ROUND(f.distance_km / 1.60934, 2) AS distance_mi,
  f.speed_kmh,
  ROUND(f.performance_seconds / 3600.0, 2) AS finish_hours
FROM marathos.gold.fct_results f
JOIN marathos.gold.dim_event e
  ON f.event_id = e.event_id
JOIN marathos.gold.dim_athlete a
  ON f.athlete_id = a.athlete_id
WHERE e.distance_unit = 'mi';

-- COMMAND ----------

SELECT 'km' AS type, COUNT(*) AS row_count
FROM marathos.gold.v_marathon_km_results
UNION ALL
SELECT 'mi' AS type, COUNT(*) AS row_count
FROM marathos.gold.v_marathon_mi_results;

-- COMMAND ----------

SELECT *
FROM marathos.gold.v_marathon_km_results
LIMIT 10;

-- COMMAND ----------

SELECT *
FROM marathos.gold.v_marathon_mi_results
LIMIT 10;
```
