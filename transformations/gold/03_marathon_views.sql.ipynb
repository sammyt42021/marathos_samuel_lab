```python
%sql
-- Databricks notebook source
-- MAGIC %md
-- MAGIC # Gold layer - Views for dashboard and Genie
-- MAGIC
-- MAGIC These views make it easier to build dashboards and ask questions in Genie.
-- MAGIC
-- MAGIC I created two views:
-- MAGIC - v_marathon_km_results
-- MAGIC - v_marathon_mi_results
-- MAGIC
-- MAGIC They both use the Gold fact and dimension tables.

-- COMMAND ----------

CREATE OR REPLACE VIEW marathos.gold.v_marathon_km_results AS
SELECT
  f.result_id,
  e.event_name,
  f.event_year,
  a.athlete_country,
  a.athlete_gender,
  f.performance_seconds,
  f.distance_km,
  f.speed_kmh,
  ROUND(f.performance_seconds / 3600.0, 2) AS finish_hours
FROM marathos.gold.fct_results f
LEFT JOIN marathos.gold.dim_event e
  ON f.event_id = e.event_id
LEFT JOIN marathos.gold.dim_athlete a
  ON f.athlete_id = a.athlete_id
WHERE f.distance_unit = 'km';

-- COMMAND ----------

CREATE OR REPLACE VIEW marathos.gold.v_marathon_mi_results AS
SELECT
  f.result_id,
  e.event_name,
  f.event_year,
  a.athlete_country,
  a.athlete_gender,
  f.performance_seconds,
  f.distance_km,
  ROUND(f.distance_km / 1.60934, 2) AS distance_mi,
  f.speed_kmh,
  ROUND(f.performance_seconds / 3600.0, 2) AS finish_hours
FROM marathos.gold.fct_results f
LEFT JOIN marathos.gold.dim_event e
  ON f.event_id = e.event_id
LEFT JOIN marathos.gold.dim_athlete a
  ON f.athlete_id = a.athlete_id
WHERE f.distance_unit = 'mi';

-- COMMAND ----------

SELECT 'km' AS type, COUNT(*) AS row_count
FROM marathos.gold.v_marathon_km_results

UNION ALL

SELECT 'mi' AS type, COUNT(*) AS row_count
FROM marathos.gold.v_marathon_mi_results;

-- COMMAND ----------

SELECT *
FROM marathos.gold.v_marathon_km_results
LIMIT 10;

-- COMMAND ----------

SELECT *
FROM marathos.gold.v_marathon_mi_results
LIMIT 10;
```
