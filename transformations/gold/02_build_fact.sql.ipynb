```python
%sql
-- Databricks notebook source
-- MAGIC %md
-- MAGIC # Gold fact table — `fct_results`

-- COMMAND ----------

CREATE OR REPLACE TABLE marathos.gold.fct_results AS
SELECT
  s.result_id,
  s.event_id,
  s.athlete_id,
  d.date_id,
  s.performance_seconds,
  s.distance_km,
  s.speed_kmh,
  s.event_finishers
FROM marathos.silver.races_obt s
LEFT JOIN marathos.gold.dim_date d
  ON s.event_date = d.event_date
 AND s.event_year = d.event_year;

-- COMMAND ----------

SELECT COUNT(*) AS fact_rows FROM marathos.gold.fct_results;
```
