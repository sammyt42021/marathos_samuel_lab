```python
%sql
-- Databricks notebook source
-- MAGIC %md
-- MAGIC # Gold layer - Build fact table
-- MAGIC
-- MAGIC This notebook creates the main fact table for marathon results.
-- MAGIC
-- MAGIC The fact table keeps measurable values:
-- MAGIC - performance seconds
-- MAGIC - distance km
-- MAGIC - speed km/h
-- MAGIC - finishers
-- MAGIC
-- MAGIC It connects to dimensions through:
-- MAGIC - event_id
-- MAGIC - athlete_id
-- MAGIC - date_id

-- COMMAND ----------

CREATE OR REPLACE TABLE marathos.gold.fct_results AS
SELECT
  s.result_id,
  s.event_id,
  s.athlete_id,

  sha2(concat_ws('|',
    cast(s.event_year as string),
    coalesce(s.event_dates, 'unknown')
  ), 256) AS date_id,

  s.event_year,
  s.performance_seconds,
  s.distance_km,
  s.distance_unit,
  s.speed_kmh,
  s.event_finishers
FROM marathos.silver.races_obt s
WHERE s.result_id IS NOT NULL
  AND s.event_id IS NOT NULL
  AND s.athlete_id IS NOT NULL
  AND s.event_year IS NOT NULL;

-- COMMAND ----------

SELECT COUNT(*) AS fact_rows
FROM marathos.gold.fct_results;

-- COMMAND ----------

SELECT *
FROM marathos.gold.fct_results
LIMIT 10;
```
