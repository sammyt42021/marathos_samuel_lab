```python
%sql
-- Databricks notebook source
-- MAGIC %md
-- MAGIC # Bronze - ingest raw race data
-- MAGIC Bronze is the raw layer.
-- MAGIC
-- MAGIC I load the original CSV file from the Unity Catalog volume into a Delta table.
-- MAGIC I rename the columns to simple snake_case names because Delta tables do not allow spaces or special characters in column names by default.

-- COMMAND ----------

CREATE OR REPLACE TABLE marathos.bronze.races_raw AS
SELECT
  `Year of event` AS year_of_event,
  `Event dates` AS event_dates,
  `Event name` AS event_name,
  `Event distance/length` AS event_distance_length,
  `Event number of finishers` AS event_number_of_finishers,
  `Athlete performance` AS athlete_performance,
  `Athlete club` AS athlete_club,
  `Athlete country` AS athlete_country,
  `Athlete year of birth` AS athlete_year_of_birth,
  `Athlete gender` AS athlete_gender,
  `Athlete age category` AS athlete_age_category,
  `Athlete average speed` AS athlete_average_speed,
  `Athlete ID` AS athlete_id_raw
FROM read_files(
  '/Volumes/marathos/default/raw/*.csv',
  format => 'csv',
  header => true,
  inferSchema => true
);

-- COMMAND ----------

SELECT COUNT(*) AS bronze_rows
FROM marathos.bronze.races_raw;

-- COMMAND ----------

SELECT *
FROM marathos.bronze.races_raw
LIMIT 10;
```
