## Cleaning decision

The Bronze table had 746,195 rows. After cleaning, the Silver table has 509,736 rows.

I removed rows that were not suitable for fair marathon analysis:
- missing event name, country, gender, distance or performance
- invalid performance time values
- unrealistic speed values above 25 km/h
- rows before 2010
- multi-day/stage events
- invalid or unsupported distance formats

This follows the lecturer's advice that it is acceptable to remove bad or out-of-scope data if the reason is clearly documented.
