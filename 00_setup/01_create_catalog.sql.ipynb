```python
%sql
-- Databricks notebook source
-- MAGIC %md
-- MAGIC # Setup - Unity Catalog
-- MAGIC This notebook creates the catalog, schemas and volume used in the Marathos medallion pipeline.
-- MAGIC
-- MAGIC Catalog: marathos
-- MAGIC Schemas: bronze, silver, gold
-- MAGIC Volume: marathos.default.raw

-- COMMAND ----------

CREATE CATALOG IF NOT EXISTS marathos;

-- COMMAND ----------

CREATE SCHEMA IF NOT EXISTS marathos.default;
CREATE SCHEMA IF NOT EXISTS marathos.bronze;
CREATE SCHEMA IF NOT EXISTS marathos.silver;
CREATE SCHEMA IF NOT EXISTS marathos.gold;

-- COMMAND ----------

CREATE VOLUME IF NOT EXISTS marathos.default.raw;

-- COMMAND ----------

SHOW SCHEMAS IN marathos;

-- COMMAND ----------

DESCRIBE VOLUME marathos.default.raw;
```
