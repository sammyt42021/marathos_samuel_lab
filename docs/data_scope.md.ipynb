# Data scope and cleaning

## Kept

- Events from **2010** onward (`utils/config.py`).
- Marathon distances: ~42 km, marathon in name, or 26.2 mi.
- Valid unit pairs:
  - `km` / `mi` distance → performance in `h`
  - `h` distance → performance in `km`

## Removed

- Distance unit `d` (days).
- Invalid event/performance combinations.
- Multi-day sub-stages (`Etappen` / `etapp` in event name).
- Speed outside 2–25 km/h when it cannot be fixed from distance and time.
- Missing event name or performance.

## IDs

- `event_id`: `dense_rank()` on `event_name`
- `athlete_id`: `dense_rank()` on country, birth year, gender, source athlete id
