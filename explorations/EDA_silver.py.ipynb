```python
# Databricks notebook source

# COMMAND ----------

from pyspark.sql import functions as F

SILVER_TABLE = "marathos.silver.races_obt"

# COMMAND ----------

silver = spark.table(SILVER_TABLE)

# COMMAND ----------

print("Silver rows:", silver.count())
print("Silver columns:", len(silver.columns))

silver.printSchema()

# COMMAND ----------

display(
    silver.describe(
        "performance_seconds",
        "distance_km",
        "speed_kmh",
        "athlete_age"
    )
)

# COMMAND ----------

display(
    silver.groupBy("event_year")
    .count()
    .orderBy("event_year")
)

# COMMAND ----------

display(
    silver.groupBy("athlete_country")
    .count()
    .orderBy(F.col("count").desc())
    .limit(20)
)

# COMMAND ----------

display(
    silver.groupBy("athlete_gender")
    .count()
    .orderBy(F.col("count").desc())
)

# COMMAND ----------

display(
    silver.groupBy("distance_unit")
    .count()
    .orderBy(F.col("count").desc())
)

# COMMAND ----------

display(
    silver.groupBy("event_name")
    .count()
    .orderBy(F.col("count").desc())
    .limit(20)
)

# COMMAND ----------

display(
    silver.select(
        "event_year",
        "event_name",
        "distance_km",
        "performance_seconds",
        "speed_kmh",
        "athlete_country",
        "athlete_gender",
        "athlete_age"
    )
    .limit(30)
)

# COMMAND ----------

import plotly.express as px

top_countries = (
    silver.groupBy("athlete_country")
    .count()
    .orderBy(F.col("count").desc())
    .limit(15)
    .toPandas()
)

fig = px.bar(
    top_countries,
    x="athlete_country",
    y="count",
    title="Top 15 countries in Silver data"
)

fig.show()

# COMMAND ----------

year_counts = (
    silver.groupBy("event_year")
    .count()
    .orderBy("event_year")
    .toPandas()
)

fig = px.line(
    year_counts,
    x="event_year",
    y="count",
    title="Number of Silver results by year"
)

fig.show()
```
